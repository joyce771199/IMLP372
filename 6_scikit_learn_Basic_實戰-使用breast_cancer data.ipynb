```python
%matplotlib inline

from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 讀取breast_cancer資料集


```python
bc = datasets.load_breast_cancer()
print(bc.DESCR)
```

* 印出key值與檔案位置
* 查看前10筆資料
* 查看資料型別
* 印出標註的樣本類別資料

```python
print(bc.keys())
print(bc['filename'])
print(bc.data[1:10])
print(type(bc.data))
print(bc.target_names)
```

```python
x = pd.DataFrame(bc.data, columns=bc['feature_names'])
x.head(10)
```

```python
print("target_names:"+str(bc["target_names"]))
```

```python
#建立Target欄位與資料
y=pd.DataFrame(bc["target"], columns=['target'])
y.head()
```

```python
#合併資料特徵欄位與目標欄位
bc_data=pd.concat([x,y],axis=1)
bc_data=bc_data[["mean radius","mean texture","mean symmetry","target"]]
bc_data.head(10)
```

## 切分訓練集與測試集
> train_test_split()

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(bc_data[["mean radius","mean texture","mean symmetry","target"]],bc_data[["target"]],test_size=0.2)
```

# Appendix 

>normalization和standardization是差不多的<br>
都是把數據進行前處理，從而使數值都落入到統一的數值範圍，從而在建模過程中，各個特徵量沒差別對待。<br> 
* normalization一般是把數據限定在需要的範圍，比如一般都是【0，1】，從而消除了數據量綱對建模的影響。<br> 
* standardization 一般是指將數據正態化，使平均值0方差為1.<br> 

因此normalization和standardization 是針對數據而言的，消除一些數值差異帶來的特種重要性偏見。<br>
經過歸一化的數據，能加快訓練速度，促進算法的收斂。

### Standardization (z-score)
    to compute the mean and standard deviation on a training set so as to be able to later reapply the same transformation on the testing set. 

## use sklearn

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)  
X_test_std = sc.fit_transform(X_test)
print(X_test_std[:10])
```

The scaler instance can then be used on new data to transform it the same way it did on the training set:

```python
print('mean of X_train_std:',np.round(X_train_std.mean(),4))
print('std of X_train_std:',X_train_std.std())
```
